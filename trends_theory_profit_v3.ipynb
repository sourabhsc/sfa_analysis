```python
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large', 'axes.linewidth' :3}
pylab.rcParams.update(params)
```

```python
from matplotlib import pyplot as plt
import numpy as np

def sigmoid(a, b,  x, tau):
    return a / (1 + a*np.exp(-(x -tau)/b))-0.5
x = np.arange(0, 25)
a = [1.1, 1.35]
b = [7, 7]
color = ['b', 'g']
ls = ['-', '-', ]
tau = [0, 0]
labels = ['Low profit', 'High profit']
k = 0
fig, ax = plt.subplots(1,1, figsize = (8, 8))

    
for a1, b1, color in zip(a, b, color,):
        
    ax.plot(x, sigmoid(a1, b1, x, tau[k]), color = color, label = labels[k], lw = 3 )#'%s, %s, %s'%(a1, b1, tau[k]), ls =ls[k])
    k = k +1

#ax[1].set_xlim(1988, 2014)
#ax[1].set_ylim(0.5, 1.3)
ax.set_xlabel('Time [in years]')
ax.set_ylabel('Industrial diversity')

ax.xaxis.set_ticks(np.arange(0,25, 2))


ax.legend()
ax.set_ylim(0,1)

fig.savefig('profi_ind_v2.png', dvi = 400)

plt.show()
    
```

```python
a = [1.2, 1.5]
b = [6, 6]
color = ['b', 'g']
ls = ['-', '-', ]
tau = [0, 0]

k = 0
labels = ['High profit', 'Low profit']

x = np.arange(0, 25)
def sigmoid(a, b,  x, tau):
    return 2 -a / (1 + a*np.exp(-(x -tau)/b)) -0.5

fig, ax = plt.subplots(1,1, figsize = (8, 8))

    
for a1, b1, color in zip(a, b, color,):
        
    ax.plot(x, sigmoid(a1, b1, x, tau[k]), color = color, label = labels[1-k], lw = 3)#'%s, %s, %s'%(a1, b1, tau[k]), ls =ls[k])
    k = k +1

#ax[1].set_xlim(1988, 2014)
#ax[1].set_ylim(0.5, 1.3)

ax.set_xlabel('Time [in years]')
ax.set_ylabel('Techinical diversity')
ax.set_ylim(0,1)

#ax.xaxis.set_ticks(np.arange(1989,2014, 4))
ax.xaxis.set_ticks(np.arange(0,25, 2))

ax.legend()
fig.savefig('profit_tech_v2.png', dvi = 400)

plt.show()
    
```
