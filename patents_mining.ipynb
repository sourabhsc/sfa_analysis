```python
import pandas as pd
import numpy as np
data = pd.read_excel('Copy of dataannual Aug1.xlsx')
```

```python
from urllib.request import urlretrieve
from urllib.parse import urlencode
import requests
from lxml import html
from bs4 import BeautifulSoup
data['patents2'] = np.zeros((len(data)))
for i in range( len(data)):
    company1 = data['companylegalname'][i].split(' ')[0]
    #replace(' ', '-')
    if len(company1)!=1:
        company3 = company1.split('.')[0]
    else:
        company3 = company1[0]
    company = company3
    year = data['datayearfiscal'][i]
    #mydict = {'Query':'ISD/1/1/1997->31/12/1997 AND AN/wipro-limited'}
    mydict = {'Query':'APD/%s0101->%s1231 AND AN/%s'%(year,year, company)}
    #mydict = {'Query':'APD/19970101->20111231 AND AN/wipro-limited'}
    
    query_string = urlencode(mydict)
    #url = 'http://patft.uspto.gov/netacgi/nph-Parser?\
    #Sect1=PTO2&Sect2=HITOFF&u=%2Fnetahtml%2FPTO%2Fsearch-adv.htm&r=0&p=1&f=S&l=50&\
    #{}%0D%0A%0D%0A&d=PTXT'.format(query_string)
    
    
    
 
    url = 'http://patft.uspto.gov/netacgi/nph-Parser?Sect1=PTO2&Sect2=HITOFF&u=%2Fnetahtml%2FPTO%2Fsearch-adv.htm&r=0&p=1&f=S&l=50&{}%0D%0A%0D%0A&d=PTXT'.format(query_string)
    r  = requests.get(url)
    html_data = r.text
    soup = BeautifulSoup(html_data)
    #print (soup)
    if soup.find('b')!=None:
        
    
        patent_string = soup.find('b').next_sibling
        num_pat = int(str(patent_string).split()[1])
        print (i, company, year, num_pat,)

        data['patents2'][i] = num_pat
    else:
        data['patents2'][i] = 0
```
